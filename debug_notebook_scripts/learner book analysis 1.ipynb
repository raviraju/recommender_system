```python
#eliminate leaners with null id
df = df[df['learner_id'].notnull()]
df.shape
```

```python
#eliminate books with null code
df = df[df['book_code'].notnull()]
df.shape
```

```python
learner_book_df = df.groupby(['learner_id','book_code']).size().reset_index().rename(columns={0:'events_count'})
```

```python
learner_book_df
```

```python
len(learner_book_df['learner_id'].unique())
```

```python
see = learner_book_df.groupby(['book_code']).agg({'learner_id': (lambda x: list(x.unique()))}).rename(columns={'learner_id':'learners'}).reset_index()
see
```

```python
len(list(see[see['book_code'] == 'BO-20170627173817922']['learners'].values)[0])
```

```python
books_stats = learner_book_df.groupby(['book_code']).agg({'learner_id': 'count'}).rename(columns={'learner_id':'no_of_learners'})
books_stats = books_stats.sort_values('no_of_learners', ascending=False).reset_index()
books_stats.to_csv('books_stats.csv')
books_stats
```

```python
learner_stats = learner_book_df.groupby(['learner_id']).agg({'book_code': 'count'}).rename(columns={'book_code':'no_of_books'})
learner_stats = learner_stats.sort_values('no_of_books', ascending=False).reset_index()
learner_stats.to_csv('learner_stats.csv')
learner_stats
```
