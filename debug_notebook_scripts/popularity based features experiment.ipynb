```python
import pandas as pd
```

```python
train_data = pd.read_csv('recommender_system/books_recommender/train_test_data/train_data.csv')
test_data = pd.read_csv('recommender_system/books_recommender/train_test_data/test_data.csv')
```

```python
user_id_col = 'learner_id'
item_id_col = 'book_code'
train_data_grouped = train_data.groupby([item_id_col])\
                                            .agg({user_id_col: 'count'})\
                                            .reset_index()
train_data_grouped.rename(columns={user_id_col:'no_of_users',
                                   item_id_col:'item_id'},
                          inplace=True)

#Sort the items based upon popularity score : no_of_users
train_data_sort = train_data_grouped.sort_values(['no_of_users', 'item_id'],
                                                 ascending=[0, 1])

#Generate a recommendation rank based upon score : no_of_users
train_data_sort['rank'] = train_data_sort['no_of_users']\
                          .rank(ascending=0, method='first')
train_data_sort.reset_index(drop=True, inplace=True)
train_data_sort
```

```python
def filter_data_frame(dataframe, feature):
    feature_vals = dataframe[feature].unique()
    filtered_dfs = dict()
    for val in feature_vals:
        filtered_dfs[val] = dataframe[dataframe[feature] == val]
    return filtered_dfs

def feature_filters(dataframe, features):
    print(features, len(dataframe))
    for feature in features:
        filtered_dfs = filter_data_frame(dataframe, feature)
        print(filtered_dfs.keys())
        for val, df in filtered_dfs.items():
            remaining_features = features[1:]
            if len(remaining_features) == 0:
                print(val)
                print(df.head())
                #input()
            else:
                feature_filters(df, remaining_features)
```
