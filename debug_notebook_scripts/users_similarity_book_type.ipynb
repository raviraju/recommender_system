```python
import pandas as pd
```

```python
df = pd.read_csv('../books_recommender/preprocessed_metadata/learner_age_item_type_stats.csv')
```

```python
df.head()
```

```python
all_learners = set(df['learner_id'].unique())
print(len(all_learners))
```

```python
media_preference = dict()
for learner_id in all_learners:
    media_preference[learner_id]={'book_close' : 0.0,
                                  'audio_close' : 0.0,
                                  'video_close' : 0.0,
                                  'age' : 0
                                 }
for index, row in df.iterrows():
    learner_id = row['learner_id']
    event_name = row['event_name']
    preference = row['percentage']
    media_preference[learner_id][event_name] = preference
    media_preference[learner_id]['age'] = row['age']


list_of_measures = []
for learner_id in media_preference:
    details = {'learner_id' : learner_id,
               'age' : media_preference[learner_id]['age'],
               'book_close' : media_preference[learner_id]['book_close'],
               'audio_close' : media_preference[learner_id]['audio_close'],
               'video_close' : media_preference[learner_id]['video_close'],
              }
    list_of_measures.append(details)
measures_df = pd.DataFrame(list_of_measures)
measures_df.set_index('learner_id', inplace=True)
measures_matrix = measures_df.as_matrix()
```

```python
measures_df.head()
```

```python
audio_users = measures_df[(measures_df['audio_close'] == 1.0) & (measures_df['book_close'] == 0)  & (measures_df['video_close'] == 0)][['audio_close', 'book_close', 'video_close']]
book_users  = measures_df[(measures_df['book_close'] == 1.0)  & (measures_df['audio_close'] == 0) & (measures_df['video_close'] == 0)][['audio_close', 'book_close', 'video_close']]
video_users = measures_df[(measures_df['video_close'] == 1.0) & (measures_df['book_close'] == 0)  & (measures_df['audio_close'] == 0)][['audio_close', 'book_close', 'video_close']]
```

```python
print("len(measures_df : all_users) : ", len(measures_df))
print("len(audio_users) : ", len(audio_users))
print("len(video_users) : ", len(video_users))
print("len(book_users) : ", len(book_users))
```

```python
# train_data_file = '../books_recommender/train_test_data/train_data.csv'
# test_data_file = '../books_recommender/train_test_data/test_data.csv'
train_data_file = '../../Dec_17/10_close_events/train_test_data/train_data.csv'
test_data_file = '../../Dec_17/10_close_events/train_test_data/test_data.csv'
train_data = pd.read_csv(train_data_file)
test_data = pd.read_csv(test_data_file)

train_users = set(train_data['learner_id'].unique())
test_users = set(test_data['learner_id'].unique())

measures_df_train = measures_df[measures_df.index.isin(train_users)]
measures_df_test = measures_df[measures_df.index.isin(test_users)]

measures_df_train_test = pd.concat([measures_df_train, measures_df_test])
len(measures_df_train_test), len(measures_df_train), len(measures_df_test)
```

```python
users = measures_df_train_test.index
users_cosine_similarity = cosine_similarity(measures_df_train_test.as_matrix())
users_cosine_similarity_df = pd.DataFrame(users_cosine_similarity,
                                  columns=users,
                                  index=users)
```
