```python
import pandas as pd
```

```python
df = pd.read_csv('learner_books_close_events.csv')
```

```python
lbdf = df.groupby('learner_id').agg({'book_code' : 'count'}).rename(columns={'book_code' : 'no_of_books'}).reset_index()
dist = lbdf['no_of_books'].describe()
no_of_books_list = [dist['min'], dist['25%'], dist['50%'], dist['75%'], dist['max']]
for no_of_books in no_of_books_list:
    print("No of Books : ", no_of_books, " No of learners : ", len(lbdf[lbdf['no_of_books'] == no_of_books]))
```

```python
min_no_of_books = 10
min_lbdf_df = lbdf[lbdf['no_of_books'] >= min_no_of_books]
print("Min no of Books : ", min_no_of_books,
      " No of learners : ",
      len(min_lbdf_df))
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.2
train_users, test_users = train_test_split(min_lbdf_df,
                                           test_size=test_size,
                                           random_state=None)
                                           #random_state=0)
data = pd.merge(min_lbdf_df, df, how='inner', on='learner_id')
train_data = pd.merge(train_users, df, how='inner', on='learner_id')
test_data = pd.merge(test_users, df, how='inner', on='learner_id')
```
